```python
from typing import Any


def user(name:str, roll:int, age:int=18, **kargs:dict[str, Any]) -> None:
    print(name)
    print(roll)
    print(age)
    for k, v in kargs.items():
        print(k, ':', v)
```

```python
user("azim", 28, friend="musfic")

user("azim", 28, 25, friend="musfic")
```

```python
class Dog():
    """simple Dog Classs"""
    def __init__(self,name,age):
        """initilize funtion"""
        self.name=name
        self.age=age
    def sit(self):
        print(self.name.title() +' is now setting')
    def roll_over(self):
        print(self.name.title()+ ' rolled over')

dog1 =Dog('dog',34)
```

```python
print(dog1)
dog1.sit()
dog1.roll_over()
```
